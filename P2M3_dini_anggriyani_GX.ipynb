````python
from great_expectations.data_context import FileDataContext
````

````python
context = FileDataContext.create(project_root_dir='./')
````

````python
# Give a name to a Datasource. This name must be unique between Datasources.
datasource_name = 'csv-data-game'
datasource = context.sources.add_pandas(datasource_name)

# Give a name to a data asset
asset_name = 'game'
path_to_data = "P2M3_dini_anggriyani_data_raw.csv"
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)

# Build batch request
batch_request = asset.build_batch_request()
````

````python
# Creat an expectation suite
expectation_suite_name = 'expectation-trip-dataset'
context.add_or_update_expectation_suite(expectation_suite_name)

# Create a validator using above expectation suite
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)

# Check the validator
validator.head()
````

# Expectation 1: to be unique

Great expectation to be unique bertujuan untuk menguji apakah benar dataset yang dimiliki memiliki nilai unik yang berbeda-beda antar barisnya. Uji coba kali ini akan ditetapkan pada kolom Rank

````python
# Expectation 1 : Kolom `Rank` harus memiliki nilai unik

validator.expect_column_values_to_be_unique('Rank')
````

# Expectation 2: to be between min_value and max_value

Great expectation to be between min_value and max_value bertujuan untuk memastikan bahwa kolom tertentu memiliki nilai antara nilai minimal dan nilai maksimal. Dataset akan diuji apakah Kolom Tahun benar memiliki nilai value minimal 1980 dan nilai maksimal 2020

````python
df.describe().T
````

````python
# Expectation 2 : Kolom `Year` nilainya harus berada di antara min. value dan max. value

validator.expect_column_values_to_be_between(
    column='Year', min_value=1980, max_value=2020
)
````

# Expectation 3: to be in set

Great expectation ini bertujuan untuk memastikan dataset mengandung nilai-nilai tertentu pada kolom yang telah ditentukan. Pada kasus ini, kolom Genre akan diuji apakah memiliki nilai-nilai yang diharapkan

````python
df.columns
````

````python
unik = df['Genre'].unique()
len(unik)
unik
````

````python
# Expectation 3 : Kolom `Genre` harus memiliki 12 hal sebagai berikut :
# Sports
# Platform
# Racing
# Role-Playing
# Puzzle
# Misc
# Shooter
# Simulation
# Action
# Fighting
# Adventure
# Strategy

validator.expect_column_values_to_be_in_set('Genre', ['Sports', 'Platform', 'Racing', 'Role-Playing', 'Puzzle', 'Misc',
       'Shooter', 'Simulation', 'Action', 'Fighting', 'Adventure',
       'Strategy'])
````

# Expectation 4: to be in type list

Great expectation to be in type list bertujuan untuk memastikan suatu kolom bertipe data tertentu. Dalam hal ini, kita akan memastikan bahwa kolom NA_Sales bertipe numerik

````python
# Expectation 4 : Kolom `NA_Sales` harus bertipe numerik (integer atau float)

validator.expect_column_values_to_be_in_type_list('NA_Sales', ['integer', 'float'])
````

# Expectation 5: to be not in set

Great expectation ini bertujuan untuk memastikan dataset tidak mengandung nilai-nilai tertentu pada kolom yang telah ditentukan. Pada kasus ini, kolom Genre diharapkan tidak mengandung nilai Drama maupun Romantic karena kedua nilai tersebut merupakan genre film, bukan game

````python
# Expectation 5 : Kolom `Genre` tidak mengandung value drama dan romantic

validator.expect_column_values_to_not_be_in_set('Genre', ['Drama', 'Romantic'])
````

# Expectation 6: mean to be between

Great expectation ini untuk memastikan bahwa kolom tertentu memiliki nilai rata-rata yang masuk ke dalam range yang diharapkan

````python
# Expectation 6 : Rata-rata kolom `Global_Sales` harus berada di rentang 0 dan 1

validator.expect_column_mean_to_be_between('Global_Sales', min_value=0, max_value=1)
````

# Expectation 7: value lengths to be between

Great expectation ini bertujuan untuk mengecek nilai kolom Publisher apakah memiliki jumlah panjang karakter sesuai dengan kriteria yang telah ditetapkan

````python
# Expectation 7 : Panjang value kolom `Publisher` harus berada antara 0 dan 50

validator.expect_column_value_lengths_to_be_between('Publisher', min_value=0, max_value=50)
````
